```python
import pandas as pd
```

```python
data = {
    "EAR": [0.30, 0.28, 0.15, 0.18, 0.25],
    "Blink": [10, 12, 3, 4, 8],
    "Yawn": [0, 0, 1, 1, 0],
    "Tilt": [5, 4, 20, 25, 10],
    "Label": ["Normal", "Normal", "Sleepy", "Sleepy", "Tired"]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.ensemble import RandomForestClassifier

# Features (input)
X = df[["EAR", "Blink", "Yawn", "Tilt"]]

# Labels (output)
y = df["Label"]

# Model
model = RandomForestClassifier()

# Training
model.fit(X, y)

print("Model Trained Successfully ✅")
```

```python
y_pred = model.predict(X)
print("Predictions:", y_pred.tolist())
```

```python
y_pred = model.predict(X)
print("Predictions:", y_pred.tolist())
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y, y_pred))
```

```python
new_data = [[0.20, 5, 1, 22]]  # Example values
prediction = model.predict(new_data)
print("Driver State:", prediction[0])
```

```python
prediction = model.predict([[0.18, 4, 1, 20]])

print("Prediction:", prediction)
```
